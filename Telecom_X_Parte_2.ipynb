# ü§ñ Modelado Predictivo

### Separacion de datos

```python
# tama√±o de base de datos
df_final.shape
```

```python
from sklearn.model_selection import train_test_split

# Definir variables predictoras (X) y variable objetivo (y)
X = df_final.drop(columns=["Evasion_1"])   # todas las variables excepto la de cancelaci√≥n
y = df_final["Evasion_1"]                  # variable objetivo (cancelaci√≥n)


```

```python
# Dividir en entrenamiento y prueba (70% entrenamiento, 30% prueba)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print("Tama√±o de entrenamiento:", X_train.shape)
print("Tama√±o de prueba:", X_test.shape)
```

### Creacion de Modelos

```python
# Normalizar para regresion Logistica (el modelo)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Crear los dos modelos
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Modelo 1: Regresi√≥n Log√≠stica (con datos normalizados)
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train_scaled, y_train)

# Modelo 2: Random Forest (sin normalizar)
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
```

### Evaluando los modelos

```python
from sklearn.metrics import accuracy_score, classification_report

# Predicciones
y_pred_log = log_reg.predict(X_test_scaled)
y_pred_rf = rf.predict(X_test)

# Evaluaci√≥n
print("üìä Regresi√≥n Log√≠stica")
print("Accuracy:", accuracy_score(y_test, y_pred_log))
print(classification_report(y_test, y_pred_log))

print("üìä Random Forest")
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

La Regresi√≥n Log√≠stica requiere normalizaci√≥n, porque de lo contrario, variables con valores grandes dominar√≠an el c√°lculo de los coeficientes y sesgar√≠an el modelo.

El Random Forest no necesita normalizaci√≥n, porque trabaja dividiendo los datos en base a reglas (√°rboles), por lo que la escala de los valores no afecta su desempe√±o.

```python
# hallando la Matriz de Confusion
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Matriz de confusi√≥n - Regresi√≥n Log√≠stica
cm_log = confusion_matrix(y_test, y_pred_log)
disp_log = ConfusionMatrixDisplay(confusion_matrix=cm_log, display_labels=[0,1])
disp_log.plot(cmap="Blues")
plt.title("Matriz de Confusi√≥n - Regresi√≥n Log√≠stica")
plt.show()

# Matriz de confusi√≥n - Random Forest
cm_rf = confusion_matrix(y_test, y_pred_rf)
disp_rf = ConfusionMatrixDisplay(confusion_matrix=cm_rf, display_labels=[0,1])
disp_rf.plot(cmap="Greens")
plt.title("Matriz de Confusi√≥n - Random Forest")
plt.show()
```

### Evaluacion de Modelos



**Mejor desempe√±o entre los modelos**

Regresi√≥n Log√≠stica fue superior: mayor exactitud (0.79 vs 0.77) y mejores m√©tricas para la clase cancelacion(evasion) (precisi√≥n, recall y F1-score).

El modelo de Regresion Logistica

- Este modelo logra un buen balance, con un recall de 0.53, lo que significa que detecta poco m√°s de la mitad de los clientes que realmente cancelan.

- Identifica mejor a los clientes que NO cancelan (clase 0) que a los que s√≠ cancelan.

**Overfitting o Underfitting ¬øLo tienen los modelos?**

Random Forest parece estar sufriendo un poco de overfitting, porque tiende a memorizar patrones de la clase mayoritaria (clientes que no cancelan) y pierde capacidad de generalizar en la minoritaria.

Regresi√≥n Log√≠stica no muestra overfitting fuerte, pero s√≠ algo de underfitting, porque su capacidad predictiva es limitada al ser un modelo lineal, y no logra capturar toda la complejidad de los datos.

**Posibles causas**

Desbalance de clases (m√°s clientes que no cancelan que clientes que s√≠). Esto afecta tanto a Regresi√≥n Log√≠stica como a Random Forest.

Variables con alta correlaci√≥n (multicolinealidad) pueden afectar especialmente a la regresi√≥n log√≠stica.

**Ajustes recomendados**

Aplicar t√©cnicas de balanceo (como SMOTE o undersampling) para mejorar recall en la clase minoritaria (cancelaci√≥n).

Afinar hiperpar√°metros de Random Forest (n_estimators, max_depth, min_samples_split) para reducir overfitting.

Probar modelos m√°s robustos como XGBoost o LightGBM, que suelen manejar mejor el desbalance.

El modelo de Regresion logistica tiene un mejor balance que el de random forest. da mayor exactitud en un 0.79.Se podria mejorar ello, evitando la multicolinealidad  dado que sigue habiendo un valor VIF en multicolinealidad muy alto que son los cargos mensuales, sin embargo se dejo esa variable para evitar tener menos datos e incurir en sesgos.
